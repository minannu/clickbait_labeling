```python
from torch.utils.data import Dataset
from transformers import CLIPProcessor, CLIPModel, BertTokenizer
from PIL import Image
import torch

class MultiModalDataset(Dataset):
    def __init__(self, dataframe, clip_processor, bert_tokenizer, transform=None):
        self.data = dataframe
        self.clip_processor = clip_processor
        self.bert_tokenizer = bert_tokenizer

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]

        # Image
        image = Image.open(row['local_directory']).convert('RGB')
        image_input = self.clip_processor(images=image, return_tensors="pt")['pixel_values'].squeeze(0)

        # Text
        text = row['title'] + " " + row['description']
        text_input = self.bert_tokenizer(text, padding='max_length', truncation=True, max_length=128, return_tensors="pt")
        input_ids = text_input['input_ids'].squeeze(0)
        attention_mask = text_input['attention_mask'].squeeze(0)

        # Label
        label = torch.tensor(row['is_clickbait'], dtype=torch.long)

        return {
            'image': image_input,
            'input_ids': input_ids,
            'attention_mask': attention_mask,
            'label': label
        }
```

```python
import torch.nn as nn
from transformers import BertModel, CLIPModel

class MultiModalClassifier(nn.Module):
    def __init__(self, text_model_name='bert-base-multilingual-cased', image_model_name='openai/clip-vit-base-patch32', hidden_dim=256, num_classes=2):
        super(MultiModalClassifier, self).__init__()
        self.text_encoder = BertModel.from_pretrained(text_model_name)
        self.image_encoder = CLIPModel.from_pretrained(image_model_name).vision_model
        
        self.image_proj = nn.Linear(768, hidden_dim)
        self.text_proj = nn.Linear(768, hidden_dim)

        self.classifier = nn.Sequential(
            nn.ReLU(),
            nn.Linear(hidden_dim * 2, num_classes)
        )

    def forward(self, image, input_ids, attention_mask):
        # Encode text
        text_output = self.text_encoder(input_ids=input_ids, attention_mask=attention_mask)
        text_feat = text_output.last_hidden_state[:, 0, :]  # [CLS] token

        # Encode image
        image_output = self.image_encoder(pixel_values=image)
        image_feat = image_output.pooler_output

        # Project and combine
        image_feat = self.image_proj(image_feat)
        text_feat = self.text_proj(text_feat)

        combined = torch.cat((image_feat, text_feat), dim=1)
        logits = self.classifier(combined)
        return logits
```

```python
from torch.utils.data import DataLoader
from transformers import BertTokenizer, CLIPProcessor
from sklearn.model_selection import train_test_split

# Load models/tokenizers
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
bert_tokenizer = BertTokenizer.from_pretrained("bert-base-multilingual-cased")

# Train-val split
train_df, val_df = train_test_split(data, test_size=0.2)

# Dataset and Dataloaders
train_dataset = MultiModalDataset(train_df, clip_processor, bert_tokenizer)
val_dataset = MultiModalDataset(val_df, clip_processor, bert_tokenizer)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64)

# Model
model = MultiModalClassifier(num_classes=2).cuda()
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)
```

```python
from tqdm import tqdm

def train_one_epoch(model, loader, optimizer, criterion):
    model.train()
    total_loss = 0

    progress_bar = tqdm(loader, desc="Training", leave=False)
    
    for batch in progress_bar:
        optimizer.zero_grad()

        inputs = {
            'image': batch['image'].cuda(),
            'input_ids': batch['input_ids'].cuda(),
            'attention_mask': batch['attention_mask'].cuda()
        }

        labels = batch['label'].cuda()
        outputs = model(**inputs)
        loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        
        avg_loss = total_loss / (progress_bar.n + 1)  # current average loss
        progress_bar.set_postfix(loss=avg_loss)

    return total_loss / len(loader)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

def evaluate(model, loader):
    model.eval()
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for batch in loader:
            inputs = {
                'image': batch['image'].cuda(),
                'input_ids': batch['input_ids'].cuda(),
                'attention_mask': batch['attention_mask'].cuda()
            }

            labels = batch['label'].cuda()
            outputs = model(**inputs)
            preds = torch.argmax(outputs, dim=1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    return np.array(all_preds), np.array(all_labels)
```

```python
from tqdm import tqdm

epochs = 5
best_val_acc = 0.0

for epoch in range(epochs):
    print(f"Epoch {epoch+1}/{epochs}")
    train_loss = train_one_epoch(model, train_loader, optimizer, criterion)
    preds, labels = evaluate(model, val_loader)

    val_acc = (preds == labels).mean()
    print(f"Train Loss: {train_loss:.4f} | Validation Accuracy: {val_acc:.4f}")

    # Save best model
    if val_acc > best_val_acc:
        torch.save(model.state_dict(), 'best_multimodal_model.pt')
        best_val_acc = val_acc
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Load best model
model.load_state_dict(torch.load('best_multimodal_model.pt'))

# Final evaluation
preds, labels = evaluate(model, val_loader)

# Report
print("Classification Report:")
print(classification_report(labels, preds))

# Confusion Matrix
cm = confusion_matrix(labels, preds)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```

```python
from PIL import Image
from transformers import BertTokenizer, CLIPProcessor
import torch

def predict_multimodal(image_path, title, description, model, tokenizer, processor, device='cuda'):
    model.eval()

    # --- 1. Process Text ---
    full_text = title + " " + description
    text_inputs = tokenizer(full_text, return_tensors="pt", padding=True, truncation=True, max_length=128)
    text_inputs = {k: v.to(device) for k, v in text_inputs.items()}

    # --- 2. Process Image ---
    image = Image.open(image_path).convert("RGB")
    image_inputs = processor(images=image, return_tensors="pt")
    image_tensor = image_inputs['pixel_values'].to(device)

    # --- 3. Forward Pass ---
    with torch.no_grad():
        logits = model(
            image=image_tensor,
            input_ids=text_inputs['input_ids'],
            attention_mask=text_inputs['attention_mask']
        )
        predicted_class = torch.argmax(logits, dim=1).item()

    return logits,predicted_class
```

```python
import os
import requests
from yt_dlp import YoutubeDL
from PIL import Image
from io import BytesIO

def get_video_info(video_id, save_dir='/kaggle/working'):
    os.makedirs(save_dir, exist_ok=True)
    video_url = f"https://www.youtube.com/watch?v={video_id}"

    try:
        ydl_opts = {
            'quiet': True,
            'no_warnings': True,
            'skip_download': True,
        }
        with YoutubeDL(ydl_opts) as ydl:
            info = ydl.extract_info(video_url, download=False)
            title = info.get('title', '')
            description = info.get('description', '')
            thumbnail_url = info.get('thumbnail', '')

            # Download thumbnail
            response = requests.get(thumbnail_url)
            response.raise_for_status()

            image = Image.open(BytesIO(response.content)).convert("RGB")
            local_path = os.path.join(save_dir, f"{video_id}.jpg")
            image.save(local_path, "JPEG")

            return {
                'local_image_path': local_path,
                'title': title,
                'description': description
            }

    except Exception as e:
        print(f"Failed to process video {video_id}: {e}")
        return None
```
